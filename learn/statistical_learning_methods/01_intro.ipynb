# 统计学习方法概论

统计学习是关于计算机基于数据构建概率统计模型并运用模型对数据进行分析与预测的一门学科。

统计学习三要素：

- 模型（假设空间）
- 策略（评价准则）
- 算法（模型学习算法）

统计学习步骤

- 有限的数据集合
- 确定包含所有可能的模型的假设空间
- 确定选择模型的准则，即学习策略
- 实现求解最优模型的算法，即学习算法
- 通过学习方法选择最优模型
- 利用最优模型对新数据进行预测分析

统计学习包括监督学习、非监督学习、半监督学习和强化学习。

分类：

- 概率模型，生成模型，存在联合概率分布
- 非概率模型，判别模型

- 线性模型
- 非线性模型

- 参数化模型
- 非参数化模型

- 在线学习
- 批量学习

- 贝叶斯学习：利用贝叶斯定理，计算给定数据条件下模型的条件概率，即后验概率
- 核方法：输入空间到特征空间的映射
